All the rows have the same 57 length as mentioned in the dataset.  

Why does it contain 57 positions?  
  
Promoters are sequence of dna/rna which tells the rna polymerase to start transcribing.  
rna polymerase binds to the promoter, and then starts transcribing.  
  
In ecoli, promoter length is around ~35. In our dataset, they have included +7(p7) sequences that will be transcribed.  
By convention, 
    position + 1 is where transcription starts (Transcription Start Site - TSS)
    positions before it are negative and
    positions after it are positive
There is no position 0

-50, ...., -1, p+1, ..., p+7

Nucleotide frequencies

```python
all_freq = pd.Series(list("".join(df['sequence']))).value_counts(normalize=True)*100
```

Nucleotide frequencies for different classes

```python
promoter_freq = pd.Series(list("".join(df[df['class']=='+']['sequence']))).value_counts(normalize=True)*100
promoter_freq
```

```python
non_promoter_freq = pd.Series(list("".join(df[df['class']=='-']['sequence']))).value_counts(normalize=True)*100
non_promoter_freq
```

```python
freq_df = pd.DataFrame({'all': all_freq, 'promoter': promoter_freq, 'non_promoter': non_promoter_freq})
freq_df.round(2)
```
