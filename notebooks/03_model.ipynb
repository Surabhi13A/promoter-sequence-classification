Here we flatten each one-hot encoded DNA sequence from a 57 × 4 matrix into a 228-dimensional feature vector.  

This is necessary because scikit-learn estimators expect the input X to have the shape (n_samples, n_features), where each row is one sample and each column is one feature.  

Our one-hot encoded data has the shape (n_samples, 57, 4), which is a 3D array. By flattening each sequence, we transform the data into the required shape (n_samples, 228).  

```python
X = X.apply(lambda seq: seq.flatten())
X.head()
```
