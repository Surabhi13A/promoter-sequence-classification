```python
mapping ={
    'a': [1, 0, 0, 0],
    'c': [0, 1, 0, 0],
    'g': [0, 0, 1, 0],
    't': [0, 0, 0, 1]
}
```

```python
def one_hot_encode(seq):
    return np.array([mapping[base] for base in seq])
```

```python
df['one_hot_sequence'] = df['sequence'].apply(one_hot_encode)
```
